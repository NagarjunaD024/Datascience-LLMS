```python
# instantiate pretrained model

import numpy as np
from pyannote.audio import Model
from scipy.spatial.distance import cdist

model = Model.from_pretrained("pyannote/wespeaker-voxceleb-resnet34-LM")
```

```python
audio_path = '/content/drive/My Drive/Colab Notebooks/DataScience+GPT/Data/audio_data/translate_german_to_french.wav'


audio_path2 = '/content/drive/My Drive/Colab Notebooks/DataScience+GPT/Data/audio_data/my_recording.wav'
```

```python
from pyannote.audio import Inference
inference = Inference(model, window="whole")
embedding1 = inference('/content/drive/My Drive/Colab Notebooks/DataScience+GPT/Data/audio_data/translate_german_to_french.wav')
embedding2 = inference('/content/drive/My Drive/Colab Notebooks/DataScience+GPT/Data/audio_data/my_recording.wav')
# `embeddingX` is (1 x D) numpy array extracted from the file as a whole.

print(f"Shape of embedding1: {embedding1.shape}")
print(f"Shape of embedding2: {embedding2.shape}")

distance = cdist(embedding1.reshape(1, -1), embedding2.reshape(1, -1), metric="cosine")[0,0]

print(f"\nCosine distance between the two recordings: {distance}")
# `distance` is a `float` describing how dissimilar speakers 1 and 2 are.
```
